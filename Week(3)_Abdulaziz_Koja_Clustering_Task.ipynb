## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
#1
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df = remove_outliers(df, col)
```

```python
#2
imputer = SimpleImputer(strategy='most_frequent')
df[df.columns] = imputer.fit_transform(df[df.columns])
```

```python
#3
leb = LabelEncoder()
for i in df.select_dtypes(include=['object']).columns:
    df[i] = leb.fit_transform(df[[i]])

sta = StandardScaler()
for i in df.select_dtypes(exclude=['object']).columns:
    print(i)
    df[i] = sta.fit_transform(df[[i]])
```

```python
#4
X = pd.DataFrame(df).to_numpy()
Z = (X - X.mean()) / X.std()

pca = PCA(n_components=2)
pca_X = pca.fit_transform(Z)
```

```python
pca_X
```

```python
#5
categorical_cols = df.select_dtypes(include=['object']).columns
one_hot_encoder = OneHotEncoder(drop='first', sparse=False)
encoded_df = pd.DataFrame(one_hot_encoder.fit_transform(df[categorical_cols]),
                          columns=one_hot_encoder.get_feature_names_out(categorical_cols))

df = df.drop(categorical_cols, axis=1).reset_index(drop=True)
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.isnull().sum()
```

## Feature Selection
Select relevant features for clustering. Explain your choice of features.


```python
df['PC1'] = pca_X[:, 0]
df['PC2'] = pca_X[:, 1]

selected_features = ['PC1', 'PC2']
df_selected = df[selected_features]
```

## Data Visualization
Visualize the data using appropriate plots to gain insights into the dataset. Using the following:
- Scatter plot of accidents based on Longitude and Latitude.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5 , color = 'Red')
plt.title('Scatter Plot of Accidents Based on Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
#1
wcss = [KMeans(n_clusters=i, init='k-means++', random_state=42).fit(df_selected).inertia_ for i in range(1, 11)]
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#2
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
df_selected['Cluster'] = kmeans.fit_predict(df_selected)

plt.scatter(df_selected['PC1'], df_selected['PC2'], c=df_selected['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
silhouette_avg = silhouette_score(df_selected[['PC1', 'PC2']], df_selected['Cluster'])
print(f'Silhouette Score: {silhouette_avg:.3f}')
```

```python
davies_bouldin = davies_bouldin_score(df_selected[['PC1', 'PC2']], df_selected['Cluster'])
print(f'Davies-Bouldin Index: {davies_bouldin:.3f}')
```

```python
inertia = kmeans.inertia_
print(f'Inertia (WCSS): {inertia:.3f}')
```

## Plot the data points with their predicted cluster center

```python
plt.scatter(df_selected['PC1'], df_selected['PC2'], c=df_selected['Cluster'], cmap='magma', alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='x', label='Centers')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
